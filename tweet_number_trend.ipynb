```python
import pandas as pd
import calendar
yrmo = {}
for yr in ['2017','2018','2019']:
    for mo in ['01','02','03','04','05','06','07','08','09','10','11','12']:
        yrmo['%s-%s',yr,calendar.month_name[int(mo)]] = 0
        
for filename in ['C:\\Users\\drill\\OneDrive\\Desktop\\143 final\\tweet_2017_to_2019_with_hashtags.csv']:
    df = pd.read_csv(filename)
    tweet_dates = df['date']

for date in tweet_dates:
    mo = (date.split(' ')[0]).split('-')[1]
    yr = (date.split(' ')[0]).split('-')[0]
    yrmo['%s-%s',yr,calendar.month_name[int(mo)]] += 1

tweet_month_df = pd.DataFrame(columns = ['time','tweets'])
    
for yr in ['2017','2018','2019']:
    for mo in ['01','02','03','04','05','06','07','08','09','10','11','12']:
        if (((int(yr)==2017) and (int(mo)>8)) or ((int(yr)==2019) and (int(mo)<7)) or int(yr)==2018):
#             print (yr,'-',calendar.month_name[int(mo)],',',yrmo['%s-%s',yr,calendar.month_name[int(mo)]])
            tweet_month_df = tweet_month_df.append({'time':yr + ' ' + calendar.month_name[int(mo)],'tweets':yrmo['%s-%s',yr,calendar.month_name[int(mo)]]},ignore_index=True)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.close('all')

fig = plt.figure(num=None, figsize=(2.56, 1.44), dpi=1000, facecolor='w', edgecolor='k')
plt.rcParams["figure.figsize"] = (25.6,14.4)
plt.rcParams.update({'font.size': 30})
tweet_month_df.plot(kind='line',x='time',y='tweets',color='blue',linewidth=6).xaxis.label.set_visible(False)

plt.ylabel('Number of Tweets')
plt.title('Number of tweets over time')

plt.annotate('Kevin Spacey', xy=(1, 4500), xytext=(0.5, 5000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Harvey Weinstein', xy=(4, 4000), xytext=(4, 4700),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Morgan Freeman', xy=(7, 2000), xytext=(7, 3000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Brett Kavanaugh', xy=(13, 5800), xytext=(14, 5500),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('R. Kelly', xy=(16, 1400), xytext=(16, 2100),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.annotate('Joe Biden', xy=(19, 1100), xytext=(19, 2000),
             arrowprops=dict(facecolor='black', shrink=0.05))

plt.savefig('C:\\Users\\drill\\OneDrive\\Desktop\\143 final\\tweets_vs_month.png')

plt.show()
```
